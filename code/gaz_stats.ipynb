```python
from collections import defaultdict
```

```python
bad_chars = "@#$%^&*+~={}[]|;"
max_chars = 50
max_tokens = 7
```

```python
def process_file(fname):
    name_count = 0
    name_chars = defaultdict(int)
    name_tokens = defaultdict(int)
    with open(fname) as infile:
        for name in (infile):
            name = name.strip()
            name_count += 1
            chars = len(name)
            tokens = len(name.split(' '))
            name_chars[chars] +=1
            name_tokens[tokens] += 1
            if tokens > max_tokens:
                print("{} tokens: {}".format(tokens, name))
            elif chars > max_chars:
                print("{} chars: {}".format(chars, name))
            else:
                for c in bad_chars:
                    if c in name:
                        print("{} bad char: {}".format(c, name))
                        continue
    return name_count, sorted(name_chars.items()), sorted(name_tokens.items())
```

```python
n, c, t = process_file('lists/eng-PER-name-wd.txt')
```
